## Calulating the log both x and y

```python
xl = np.log(x)
yl = np.log(y)
print(xl)
print(yl)
```

## Defining the functions

### Function from sum of all elements

```python
 def sum(list1):
   total = 0
   ele = 0 
   for ele in range(0, len(list1)):
     total = total + list1[ele]
   return total
```

### Function for sum of squared elements

```python
 def sum2(list1):
   total = 0
   ele = 0 
   for ele in range(0, len(list1)):
     total = total + list1[ele]**2
   return total
```

### Function from sum of product of x and y elements 

```python
def sum_xy(list1,list2):
  total = 0
  ele = 0 
  for ele in range(0, len(list1)):
    total = total + list1[ele] * list2[ele]
  return total
```

## Calculating sumx, sumx2, sumxy and sumy



```python
if k==1 or k==2:
  sumx  = sum(x)
  sumx2 = sum2(x)
  sumxy = sum_xy(x,yl)
  sumy  = sum(yl)
elif k==3:
  sumx  = sum(xl)
  sumx2 = sum2(xl)
  sumxy = sum_xy(xl,yl)
  sumy  = sum(yl)

print(f"sumx = {sumx}")
print(f"sumx2 = {sumx2}")
print(f"sumy = {sumy}")
print(f"sumxy = {sumxy}")
```

## Calculating the value of A and B

```python
n = len(x)
b = (sumx*sumy-n*sumxy)/(sumx*sumx-n*sumx2)
a = (sumy-b*sumx)/n
print("A =",a)
print("B =",b)
```

## Calculataing the value of a and b

```python
if k==1 or k==3:
  a = math.exp(a)
  b = b
elif k==2:
  a = math.exp(a)
  b = math.exp(b)
print("a =",a)
print("b =",b)
```

## Function to calculate the value of y 

```python
def f(a,b,x):
  if k==1:
    return a*(math.exp(b*x))
  elif k==2:
    return a*(b**x) 
  elif k==3:
    return a*(x**b)
f2 = np.vectorize(f)
```

```python
ylist = f2(a,b,xlist)
```
